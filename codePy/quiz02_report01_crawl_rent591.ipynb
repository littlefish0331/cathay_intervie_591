# 完整執行 - taipei

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time
import random
import re
import json
```

## setting before crawl

```python
url = 'https://rent.591.com.tw'
agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36'

header = {
    'Referer': url,
    'User-Agent': agent,
}
```

## token

and total # of data

```python
session = requests.session()
session.headers = header
source_page = session.get(url+'/?kind=0&region=1')
soup = BeautifulSoup(source_page.text, 'lxml')
```

```python
token = soup.find('meta', {'name':'csrf-token'}).get('content')
print(token)

total_n_text = soup.find('div', {'class':'pull-left hasData'}).text
total_n = int(re.sub(r'[\u4e00-\u9fff ,]+', '', total_n_text))

len(soup.find_all('li', {'class':'pull-left infoContent'})) # 可以知道一頁有30筆M
total_n = int(np.ceil(total_n/30)) #一頁30筆數據，無條件進位，看總共需要爬幾頁
```

## let's start

```python
%%time

n = 0
results_data = []
while n<(total_n):
    source_page = session.get(url+f'/home/search/rsList?kind=0&regionid=1&firstRow={30*n}',
                    headers={
                        'Referer': url,
                        'User-Agent': agent,
                        'X-CSRF-TOKEN':token
                    }).json()

    results_data = results_data + source_page['data']['data']
    
    print(f'Complete {n+1} th page and remaining page = {total_n - (n+1)}')
    n = n+1
    time.sleep(random.uniform(1, 2))
```

## save json

```python
path_json = os.path.join(path_result, './results_taipai.json')
with open(path_json, 'w', encoding='utf-8') as jsonfile:
    json.dump(results_data, jsonfile, ensure_ascii = False)
```

```python
session.close()
```
